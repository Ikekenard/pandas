## Manipulando os dados


##### `Criando colunas numéricas`

```python
df.head(6)
```

```python
df['Valor_mes'] = df['Valor'] + df['Condominio']
df.head()
```

```python
df['Valor_ano'] = 12 * (df['Valor_mes']) + df['IPTU']
df.head()
```

##### `Criando Coluna Categórica`

```python
df['Descrição'] = df['Tipo'] + ' em ' + df['Bairro'] + ' com ' + \
                                        df['Quartos'].astype(str) + 'quarto(s)' + \
                                        ' e ' + df['Vagas'].astype(str) + ' vaga(s) p/carro'
df.head()
```

```python
df['Descricao'] = df['Tipo'] + ' em ' + df['Bairro'] + ' com ' + \
                                        df['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + df['Vagas'].astype(str) + ' vaga(s) de garagem.'
df.head()
```

```python
df['Possui_suite'] = df['Suites'].apply(lambda x:'Sim' if x > 0 else 'Não')
df.head()
```
